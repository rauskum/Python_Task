# 1. Write a Python program to find a target values in a list using linear search with following steps: 
a. Initialize the list to store the input elements.

b. Initialize found-False.

C. Enter the item to be searched (match_item) .

d. For each element in the list

1. if match _item = value

a.return match_item's position.

e. If the match_item is not in the list, display an error message that the item is not found in the list.

```python
def linear_search(lst, match_item):
    found = False
    
    for i in range(len(lst)):
        if lst[i] == match_item:
            found = True
            return i
    
    if not found:
        print("Item not found in the list")

# Initialize the list
lst = []

# Take input from user for the elements of the list
n = int(input("Enter the number of elements in the list: "))
for i in range(n):
    element = int(input("Enter element {}: ".format(i+1)))
    lst.append(element)

# Take input from user for the item to be searched
match_item = int(input("Enter the item to be searched: "))

# Call the linear_search function
result = linear_search(lst, match_item)

# Print the result
if result is not None:
    print("Item found at position", result)
```

# 2. Write a Python program to implement binary search to find the target values from the list: 
a. Create a separate function to do binary search.

b. Get the number of inputs from the user.

C. Store the inputs individually in a list. d. In binary search function at first sort the list in order to start the search from middle of the list.

e. Compare the middle element to right and left elements to search target element. 
f. If greater, move to right of list or else move to another side of the list.

g. Print the result along with the position of the element.

```python
def binary_search(lst, match_item):
    lst.sort()  # Sort the list in ascending order
    
    low = 0
    high = len(lst) - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        if lst[mid] == match_item:
            return mid
        elif lst[mid] < match_item:
            low = mid + 1
        else:
            high = mid - 1
    
    return None

# Take input from user for the number of elements in the list
n = int(input("Enter the number of elements in the list: "))

# Initialize an empty list
lst = []

# Take input from user for the elements of the list
for i in range(n):
    element = int(input("Enter element {}: ".format(i+1)))
    lst.append(element)

# Take input from user for the item to be searched
match_item = int(input("Enter the item to be searched: "))

# Call the binary_search function
result = binary_search(lst, match_item)

# Print the result
if result is not None:
    print("Item found at position", result)
else:
    print("Item not found in the list")
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm:
 a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted

list.

c. In the second iteration, find the smallest element in the unsorted list and place it in the

correct position by comparing with the element in the sorted list.

d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list.

e. This process continues till the unsorted list becomes empty.

f. Display the sorted list.


```python
def selection_sort(lst):
    n = len(lst)
    
    for i in range(n-1):
        min_index = i
        
        for j in range(i+1, n):
            if lst[j] < lst[min_index]:
                min_index = j
        
        lst[i], lst[min_index] = lst[min_index], lst[i]
    
    return lst

# Take input from user for the number of elements in the list
n = int(input("Enter the number of elements in the list: "))

# Initialize an empty list
lst = []

# Take input from user for the elements of the list
for i in range(n):
    element = int(input("Enter element {}: ".format(i+1)))
    lst.append(element)

# Call the selection_sort function
sorted_lst = selection_sort(lst)

# Print the sorted list
print("Sorted list:", sorted_lst)
```

# 4. Write a Python program for sorting a list of elements using insertion sort algorithm: 
a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, take the first element in the unsorted list and insert it in Sorted list. 
C. In the second iteration, take the second element in the given list and compare with the element in the sorted sub list and place it in the correct position. 
d. In the third iteration, take the third element in the given list and compare with the

e.elements in the sorted sub list and place the elements in the correct position. This process continues until the last element is inserted in the sorted sub list.

f. Display the sorted elements.

```python
def insertion_sort(lst):
    n = len(lst)
    
    for i in range(1, n):
        key = lst[i]
        j = i - 1
        
        while j >= 0 and lst[j] > key:
            lst[j + 1] = lst[j]
            j -= 1
        
        lst[j + 1] = key
    
    return lst

# Take input from user for the number of elements in the list
n = int(input("Enter the number of elements in the list: "))

# Initialize an empty list
lst = []

# Take input from user for the elements of the list
for i in range(n):
    element = int(input("Enter element {}: ".format(i+1)))
    lst.append(element)

# Call the insertion_sort function
sorted_lst = insertion_sort(lst)

# Print the sorted list
print("Sorted list:", sorted_lst)
```

# 5 5. Write a Python program that performs merge sort on a list of numbers: 
a. Divide: If the given array has zero or one element, return.
1. Otherwise
ii. Divide the input list in to two halves each containing half of the elements. i.e. left half and right half. 
b. Conquer: Recursively sort the two lists (left half and right half).
       a. Call the merge sort on left half. 
       b. Call the merge sort on right half.

c. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return
    
    # Divide the input list into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    # Recursively sort the left and right halves
    merge_sort(left_half)
    merge_sort(right_half)
    
    # Merge the two sorted halves back into a single sorted list
    merge(arr, left_half, right_half)

def merge(arr, left_half, right_half):
    i = j = k = 0
    
    # Compare and merge elements from left and right halves into the original list
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
        else:
            arr[k] = right_half[j]
            j += 1
        k += 1
    
    # Check if there are any remaining elements in left_half or right_half
    while i < len(left_half):
        arr[k] = left_half[i]
        i += 1
        k += 1
    
    while j < len(right_half):
        arr[k] = right_half[j]
        j += 1
        k += 1

# Input list of numbers
input_list = [22, 15, 28, 35, 52, 18]

# Call the merge_sort function to sort the list
merge_sort(input_list)

# Display the sorted list
print("Sorted List:")
for num in input_list:
    print(num, end=" ")
```

# 7. Write a python program to implement the various operations for Stack ADT 
i.) Push 
ii.) Pop 
iii.) Display.

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, data):
        self.stack.append(data)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return "Stack is empty"

    def display(self):
        if not self.is_empty():
            for i in range(len(self.stack) - 1, -1, -1):
                print(self.stack[i])
        else:
            print("Stack is empty")

    def is_empty(self):
        return len(self.stack) == 0

if __name__ == '__main__':
    stack = Stack()

    # Push elements onto the stack
    stack.push(3)
    stack.push(1)
    stack.push(5)
    stack.push(7)
    stack.push(9)

    # Display the stack
    print("Stack:")
    stack.display()

    # Pop an element from the stack
    popped_element = stack.pop()
    print("Popped Element:", popped_element)

    # Display the stack after popping an element
    print("Stack after popping an element:")
    stack.display()
```

# 8. Write a python script to implement the various operations for Queue ADT
i.) Insert
ii.) Delete
iii.) Display. 


```python
class Queue:
    def __init__(self):
        self.queue = []

    def insert(self, data):
        self.queue.append(data)

    def delete(self):
        if not self.is_empty():
            return self.queue.pop(0)
        else:
            return "Queue is empty"

    def display(self):
        if not self.is_empty():
            for item in self.queue:
                print(item)
        else:
            print("Queue is empty")

    def is_empty(self):
        return len(self.queue) == 0

if __name__ == '__main__':
    queue = Queue()

    # Insert elements into the queue
    queue.insert(3)
    queue.insert(2)
    queue.insert(7)
    queue.insert(5)
    queue.insert(9)

    # Display the queue
    print("Queue:")
    queue.display()

    # Delete an element from the queue
    deleted_element = queue.delete()
    print("Deleted Element:", deleted_element)

    # Display the queue after deleting an element
    print("Queue after deleting an element:")
    queue.display()
```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack

a. A/B^C+D*E-F*G

b. (B^2-4*A*C)^(1/2) (100)

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, data):
        self.stack.append(data)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return None

    def is_empty(self):
        return len(self.stack) == 0

    def precedence(self, operator):
        if operator == '^':
            return 3
        elif operator == '*' or operator == '/':
            return 2
        elif operator == '+' or operator == '-':
            return 1
        else:
            return 0

    def infix_to_postfix(self, infix_expression):
        postfix_expression = ""
        for char in infix_expression:
            if char.isalnum():
                postfix_expression += char
            elif char == '(':
                self.push(char)
            elif char == ')':
                while not self.is_empty() and self.peek() != '(':
                    postfix_expression += self.pop()
                if not self.is_empty() and self.peek() != '(':
                    return "Invalid Expression"
                else:
                    self.pop()
            else:
                while not self.is_empty() and self.precedence(char) <= self.precedence(self.peek()):
                    postfix_expression += self.pop()
                self.push(char)

        while not self.is_empty():
            postfix_expression += self.pop()

        return postfix_expression

if __name__ == '__main__':
    stack = Stack()

    # Convert infix expression to postfix
    infix_expression = "A/B^C+D*E-F*G"
    postfix_expression = stack.infix_to_postfix(infix_expression)
    print("Infix Expression:", infix_expression)
    print("Postfix Expression:", postfix_expression)

    # Convert infix expression to postfix
    infix_expression = "(B^2-4*A*C)^(1/2) (100)"
    postfix_expression = stack.infix_to_postfix(infix_expression)
    print("Infix Expression:", infix_expression)
    print("Postfix Expression:", postfix_expression)
```
